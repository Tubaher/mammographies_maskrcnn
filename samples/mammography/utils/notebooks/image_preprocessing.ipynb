```
path_annotations = 'stuff/dataset/annotations/complete_annotations.csv'
path_images = 'stuff/dataset/images/CBIS-DDSM/'
```

```
annotations = pandas.read_csv(path_annotations)
```

```
annotations.head()
```

```
for index, row in annotations.iterrows():
  print(row[3])
```

```
for index,row in annotations.iterrows():
  src_img = cv2.imread(path_images + row[1])
  src_img = cv2.resize(src_img, (1920, 2944))
  src_crop = cv2.imread(path_images + row[2]) # resize(src_img, (2944,1920))
  src_mask = cv2.imread(path_images + row[3])
  src_mask = cv2.resize(src_mask,(1920, 2944))

  if 'LEFT' in row[1]: 
    src_img = cv2.flip(src_img, 1)

  if 'LEFT' in row[2]:
    src_crop = cv2.flip(src_crop, 1)

  if 'LEFT' in row[3]:
    src_mask = cv2.flip(src_mask, 1)

  cv2.imwrite(path_images + row[1], src_img)
  cv2.imwrite(path_images + row[2], src_crop)
  cv2.imwrite(path_images + row[3], src_mask)
```
